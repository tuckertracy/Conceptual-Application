The goal of this notebook is to explore how to make visualizations with Python and matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in the data
original_data = pd.read_csv("jobs_in_data copy.csv")

# get the head of the data
original_data.head()
```

```python
# store the count of all the categories in the experience level column
experience_group = original_data.groupby('experience_level').size()
experience_group.head()
```

```python
# make a pie chart of the experience levels
plt.pie(experience_group)
plt.show()
```

```python
# you can change color of the labels using
plt.rcParams['text.color']='black'

# set font size of text to 18
plt.rcParams['font.size'] = 12

# create a pie chart
plt.pie(experience_group,labels=experience_group.index,
        autopct='%1.0f%%',
        colors=['#8000E0','#95A128','#CBE100','#632E8B'])

# show the pie chart
plt.show()
```

Let's make a histogram to analyze the distribution of salaries. Histograms help describe the deistribution of the data you are analyzing. You should explore the data before doing any analysis, exploration can be done by creating visuals such as a histogram.

```python
# histogram of salary
plt.hist(original_data['salary'])
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')
plt.show()
```

Make subplots using pie charts.

```python
fig, axs = plt.subplots(2, figsize=(8, 8))

experienceGroup = original_data.groupby('experience_level').size()

axs[0].pie(x=experienceGroup, 
           labels=experienceGroup.index, 
           autopct = '%1.0f%%')
axs[0].text = 12

settingGroup = original_data.groupby('work_setting').size()

axs[1].pie(x=settingGroup, 
           labels=settingGroup.index, 
           autopct = '%1.0f%%')
axs[1].text = 12

plt.show()
```
