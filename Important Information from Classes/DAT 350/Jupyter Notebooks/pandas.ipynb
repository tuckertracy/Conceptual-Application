This notebook explores how we can use pandas library.

```python
# import the pandas library

import pandas as pd
```

Making an example DataFrame.

```python
# creating a dictionary (every key must be unique in a dictionary)

student_dict = {
    "Austin":["Accounting","Wings","Country"],
    "Caleb":["Finance","Cheeseburger","Country"],
    "Sean":["Data Science","Rice","Pop"],
}

student_dict
```

```python
# turn the dictionary into a DataFram
student_df = pd.DataFrame.from_dict(student_dict)
student_df
```

```python
# orient changes the orintatation of the data frame (orient is by default columns)
# each key is it's own entry, so setting orient to index changes the key to a row

student_df = pd.DataFrame.from_dict(student_dict,orient='index',columns=["Major","Fav Food","Fav Music"])
student_df
```

How to access the DataFrame.

```python
# access one entry
student_df.loc["Austin"]

# access entry and one column
student_df.loc["Austin","Fav Food"]

# access all entries on one column
student_df.loc[:,"Fav Music"]
```

How to delete columns using pandas.

```python
# axis = 0 means to delete rows, axis = 1 means to delete columns
cleaned_wine_df = wine_df.drop(["Appellation","Secondary Grape Varieties"],axis=1)
cleaned_wine_df.head()
```

Pandas efficency comes from the use of vectorization. Vectorization is similar to linear algebra when you mulitply two vectors with each other. (Pandas library was written in C) 

Get the mean of the price column.

```python
# problem is price is a string
# cleaned_wine_df["Price"].mean()
```

```python
# to replace characters in a string
# code on left of equal sign is what you are saving the changes you make on the right side
cleaned_wine_df["Price"] = cleaned_wine_df["Price"].str.replace('£','')
cleaned_wine_df["Price"] = cleaned_wine_df["Price"].str.replace(" per bottle",'')
cleaned_wine_df["Price"] = cleaned_wine_df["Price"].str.replace(" per case",'') # this could have thrown off the average 'per bottle'
cleaned_wine_df["Price"] = cleaned_wine_df["Price"].str.replace(" each",'')

cleaned_wine_df.head()
```

```python
# convert data type 
cleaned_wine_df["Price"] = cleaned_wine_df["Price"].astype(float)

# get the mean
cleaned_wine_df["Price"].mean()
```

```python
cleaned_wine_df['Description'].str.contains('award')
```

```python
cleaned_wine_df['Description'] = cleaned_wine_df['Description'].str.lower()
```

```python
# na=False means if there is an NA then treat it as False
award_df = cleaned_wine_df.loc[cleaned_wine_df['Description'].str.contains('award',na=False)] 

award_df['Price'].mean()
```

```python
# na=False means if there is an NA then treat it as False
fruity_df = cleaned_wine_df.loc[cleaned_wine_df['Description'].str.contains('fruity',na=False)]

fruity_df['Price'].mean()
```

```python
fruity_award_df = cleaned_wine_df.loc[cleaned_wine_df['Description'].str.contains('fruity',na=False) & 
                                      cleaned_wine_df['Description'].str.contains('award',na=False)]
```

```python
fruity_award_df['Price'].mean()
```

```python
cleaned_wine_df['Characteristics'] = cleaned_wine_df['Characteristics'].str.lower()

miles_df = cleaned_wine_df.loc[cleaned_wine_df['Characteristics'].str.contains('tropical',na=False)]

miles_df.head()

miles_df['Price'].mean()
```
