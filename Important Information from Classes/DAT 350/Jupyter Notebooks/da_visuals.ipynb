The goal of this notebook is to make some visualizations presenting information from a dataset on high school scores and factors that may impact those scores.

1. Import pandas and matplotlib.pyplot

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

2. Load in the student performance data set as a pandas DataFrame

```python
# read in the csv file
original_data = pd.read_csv("StudentsPerformance.csv")

# get head of student performance
original_data.head()
```

3. Generate pie charts to explore data set

3.1 Pie chart exploring gender

```python
# get the count of each entry that is of each type of gender
gender_group = original_data.groupby('gender').size()

# create the pie chart
plt.pie(gender_group)
plt.show()
```

```python
# make the index of gender_group into a list
gender_indices = gender_group.index.tolist()

# change the values of the list
gender_indices[0:2] = ['Female','Male']
gender_indices
```

3.2 Improve the pie chart

```python
# change font size to 12
plt.rcParams['font.size'] = 12

# add labels, percentages, and colors to pie chart
plt.pie(gender_group,
        labels=gender_indices,
        autopct='%1.0f%%',
        colors=['#5D10F0','#F0D100'])

# add a title to the pie chart
plt.title('Makeup of Gender')

# show the plot
plt.show()
```

3.3 Pie chart above for 'parent level of education'

```python
# get the size of each group
parent_education = original_data.groupby('parental level of education').size()

# create a list of better labels for the pie chart
parent_education_indices = parent_education.index.tolist()
parent_education_indices[0:6] = ['Associate\'s Degree',
                                 'Bachelor\'s Degree',
                                 'High School',
                                 'Master\'s Degree',
                                 'Some College',
                                 'Some High School']

# add labels, percentages, and colors to pie chart
plt.pie(parent_education,
        labels=parent_education_indices,
        autopct='%1.0f%%',
        colors=['#E09100','#285EA1','#0065E1','#8B6B2E','#304661','#A0E000'])


# add a title to the pie chart
plt.title('Makeup of Parental Education')

# show the plot
plt.show()
```

3.4 Pie chart for 'lunch' 

```python
# get the size of each group
lunch = original_data.groupby('lunch').size()

# create a list of better labels for the pie chart
lunch_indices= lunch.index.tolist()
lunch_indices[0:2] = ['Free/Reduced','Standard']

# add labels, percentages, and colors to pie chart
plt.pie(lunch,
        labels=lunch_indices,
        autopct='%1.0f%%',
        colors=['#E10101','#00E040'])


# add a title to the pie chart
plt.title('Makeup of Lunch Options')

# show the plot
plt.show()
```

3.5 Pie chart for test preparation course

```python
# get the size of each group
test_prep_course = original_data.groupby('test preparation course').size()

# create a list of better labels for the pie chart
test_prep_indices= test_prep_course.index.tolist()
test_prep_indices[0:2] = ['Completed','None']

# add labels, percentages, and colors to pie chart
plt.pie(test_prep_course,
        labels=test_prep_indices,
        autopct='%1.0f%%',
        colors=['#0065DB','#00DCC6'])


# add a title to the pie chart
plt.title('Makeup of Test Preparation')

# show the plot
plt.show()
```

4. Create histograms to explore the distributions of the math, writing, and reading scores.

```python
# histogram of math scores
plt.hist(original_data['math score'], edgecolor = 'black')

# add a x-label
plt.xlabel('Math Score')

# add a y-label
plt.ylabel('Frequency')

# add a title
plt.title('Distribution of Math Scores')

# show the plot
plt.show()
```

```python
# histogram of writing scores
plt.hist(original_data['writing score'],color='red', edgecolor = 'black')


# add a x-label
plt.xlabel('Writing Score')

# add a y-label
plt.ylabel('Frequency')

# add a title
plt.title('Distribution of Writing Scores')

# show the plot
plt.show()
```

```python
# histogram of reading scores
plt.hist(original_data['reading score'],color='green', edgecolor = 'black')


# add a x-label
plt.xlabel('Reading Score')

# add a y-label
plt.ylabel('Frequency')

# add a title
plt.title('Distribution of Reading Scores')

# show the plot
plt.show()
```

5. Overlap the three histograms to compare the distributions

```python
# histogram of math scores
plt.hist(original_data['math score'],alpha = 0.5, label='Math Scores')

# histogram of writing scores
plt.hist(original_data['writing score'],color='red', alpha = 0.5, label='Writing Scores')

# histogram of reading scores
plt.hist(original_data['reading score'],color='green', alpha = 0.5, label='Reading Scores')

# add a legend
plt.legend(loc='upper left')

# show the plot
plt.show()
```

6. Create scatterplot to observe any correlations between the three variable 

```python
# create objects for each of the scores
math = original_data['math score']
reading = original_data['reading score']
writing = original_data['writing score']
```

```python
# initialize the subplots
fig,(axs1,axs2,axs3) = plt.subplots(3,figsize=(15,15))

# add a title
fig.suptitle('Scatter Plots to Observe Correlations Among Testing Scores')

# make plot 1
axs1.scatter(math,reading,edgecolor='black')
axs1.set_xlabel('Math Scores')
axs1.set_ylabel('Reading Scores')

# make plot 2
axs2.scatter(reading, writing,edgecolor='black',color='red')
axs2.set_xlabel('Reading Scores')
axs2.set_ylabel('Writing Scores')

# make plot 3
axs3.scatter(math,writing,edgecolor='black',color='green')
axs3.set_xlabel('Math Scores')
axs3.set_ylabel('Writing Scores')

# show the figure
plt.show()
```

7. Now that you have the subplots, do you think there is a correlation between any of the scores? Which one do you think has the highest correlation based on the scatter plots? Why?

I think there are correlations between all of the scores. There is a visual trend upwards and to the right in each of the sub plots. Indicating that as the score on the x-axis increases so does the score on the y-axis. However, some correlations look higher than others. 

In particular, the scatter plot visualising the relationship between reading and writing scores has less variation than the other sub plots. Indicating that the correlation between reading and writing is higher than the correlation between math and reading and the correlation between math and writing. 

I think reading and writing would have the highest correlation because in both subject you can learn one by doing the other. As you read, you learn from reading how to write with more skill. Similarly, as you write it is typical to read what you have written which helps you practice your reading skills.

8. Calculate the correlation coefficients

```python
correlation_coefficients = {}
correlation_coefficients['Math & Reading'] = original_data['math score'].corr(original_data['reading score'])
correlation_coefficients['Math & Writing'] = original_data['math score'].corr(original_data['writing score'])
correlation_coefficients['Reading & Writing'] = original_data['reading score'].corr(original_data['writing score'])
correlation_coefficients
```

9. Which pair has the highest correlation (not including the pairs that go along the diagonal)? Does this seems consistent with what you saw with your charts?

The pair with the highest correlation is reading and writing, with a coefficient of ~0.95; this finding seems consistent with the scatter plot visualizations found above.
